# **Remove Punctuation**

```
punc = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
 
# Removing punctuations in string
# Using loop + punctuation string
def remove_punct(text):
    for i in text:
        if i in punc:
            text = text.replace(i, "")
    return text
```
